```python
import pandas as pd
```

```python
ideal = pd.read_csv("data.csv")
ideal
```

```python
ideal = (ideal.groupby((ideal.materials != ideal.materials.shift()).cumsum()).mean().reset_index(drop=True))
ideal
```

```python
ideal = ideal.drop(columns = ["distance", "mode"])
ideal
```

```python
ideal = ideal.astype({"materials": int})
ideal
```

```python
mid = ideal["spectrum_165"].to_numpy()
mid
```

```python
col = ["Cardboard", "Ceramic", "Concrete", "Foam", "Glass", "Plastic", "Rock", "Stainless Steel", "Wood"]
```

```python
ideal.insert(1, "names", col)
```

```python
ideal.to_csv("ideal.csv", index=False)
```

```python
dae_label = ideal.copy()
ldeal_base = ideal.copy()
for i in range(139):
    dae_label = pd.concat([dae_label, ldeal_base])
```

```python
dae_label = dae_label.sort_values(by=['materials'])
```

```python
dae_label.to_csv("dae_label.csv", index = False)
```
