```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from keras.models import Sequential, Model, load_model
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop, SGD, Adam
from sklearn.model_selection import train_test_split
import keras
```

```python
# median of 331 is 166
data  =  pd.read_csv("data.csv")
data
```

```python
e_data = data.copy()
```

```python
test = e_data[(e_data["names"]=="Cardboard1" )|(e_data["names"]=="Ceramic1" )|(e_data["names"]=="Glass1") |(e_data["names"]=="Plastic1") |(e_data["names"]=="Rock1")].reset_index(drop=True)

test
```

```python
test["names"].drop_duplicates()
```

```python
test_y = test["materials"]
test_x = test.drop(columns = ["names", "color", "distance", "mode", "materials"])
```

```python
train_data = e_data[(e_data["names"]!="Cardboard1" )&(e_data["names"]!="Ceramic1" )&(e_data["names"]!="Glass1")&(e_data["names"]!="Plastic1") &(e_data["names"]!="Rock1")].reset_index(drop=True)
train_data
```

```python
train_data["names"].drop_duplicates()
```

```python
y = train_data["materials"]
x = train_data.drop(columns = ["names", "color", "distance", "mode", "materials"])
```

```python
train_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
BATCH_SIZE = 32
EPOCHS = 500
def mlp():
    model = Sequential()
    model.add(Dense(331, activation='relu', input_shape=(331,)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(9, activation='softmax'))
    return model
```

```python
x_tra, x_val, y_tra, y_val = train_test_split(x, y, test_size=0.2, random_state=50)
y_tra = keras.utils.to_categorical(y_tra, 9)
y_val = keras.utils.to_categorical(y_val, 9)
```

```python

model = mlp()
adam = Adam(lr=0.0001)
model.compile(loss='categorical_crossentropy',
              optimizer=adam,
              metrics=['accuracy'])

history = model.fit(x_tra, y_tra,
                    batch_size=BATCH_SIZE,
                    epochs=EPOCHS,
                    verbose=1,
                    validation_data=(x_val, y_val)
                    )
```

```python
test_y = keras.utils.to_categorical(test_y, 9)
```

```python
score = model.evaluate(x_val, y_val, verbose=0)
score
```

```python
score = model.evaluate(test_x, test_y, verbose=0)
score
```

```python
pred = model.predict(test_x) 
pred = np.argmax(pred, axis = 1)
label = np.argmax(test_y,axis = 1)

print(pred) 
print(label)
count = 0 

for i in range(len(pred)):
    if(pred[i]==label[i]):
        count += 1
print(count/len(pred))
```
