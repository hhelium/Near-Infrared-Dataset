```python
canonical = pd.read_csv("data.csv")
canonical = canonical.drop(columns = ["names", "color", "distance", "mode"])
canonical
```

```python
# material 0 Cardboard
def moveData(material):
    cardboard = canonical[canonical["materials"]==material]
    cardboard_label = cardboard.pop("materials")

    cardboard1 = cardboard[cardboard["spectrum_165"]>= mid[material]]
    for i in range(330):
        s = "spectrum_" + str(i)
        cardboard1.loc[:,s] = cardboard1.loc[:,s] - (cardboard1.loc[:,"spectrum_165"] - mid[material])


    cardboard2 = cardboard[cardboard["spectrum_165"]< mid[material]]
    for i in range(330):
        s = "spectrum_" + str(i)
        cardboard2.loc[:,s] = cardboard2.loc[:,s] + (mid[material] - cardboard2.loc[:,"spectrum_165"])

    cardboard = pd.concat([cardboard1, cardboard2])
    cardboard = pd.concat([cardboard_label, cardboard], axis = 1)
    return cardboard
```

```python
# material 1 Ceramic

ceramic = canonical[canonical["materials"]==1]
ceramic_label = ceramic.pop("materials")

ceramic1 = ceramic[ceramic["spectrum_165"]>= mid[1]]
for i in range(330):
    s = "spectrum_" + str(i)
    ceramic1.loc[:,s] = ceramic1.loc[:,s] - (ceramic1.loc[:,"spectrum_165"] - mid[1])


ceramic2 = ceramic[ceramic["spectrum_165"]< mid[1]]
for i in range(330):
    s = "spectrum_" + str(i)
    ceramic2.loc[:,s] = ceramic2.loc[:,s] + (mid[1] - ceramic2.loc[:,"spectrum_165"])

ceramic = pd.concat([ceramic1, ceramic2])
ceramic = pd.concat([ceramic_label, ceramic], axis = 1)
```

```python
c_data = moveData(0)
for i in range(1,9):
    c_data = pd.concat([c_data, moveData(i)])
```
